## Trends of Skills Required in Job Postings

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns

# Getting in the Dataset
df = pd.read_csv('D:\Python Analysis\Python Analytics\data_jobs.csv')

# Minimal cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

#### Filtering the dataset for Data Analyst role

```python
df = df[df['job_title_short'] == 'Data Analyst']
```

#### Adding MonthNum and MonthName 

```python
df = df.copy()
df['monthNum'] = df['job_posted_date'].dt.month
df['monthName'] = df['job_posted_date'].dt.strftime('%b')
df
```

#### Exploding the job skills column

```python
df_exploded = df.explode('job_skills')
```

#### Grouping the Skills by Month

```python
df_trendPivot = df_exploded.pivot_table(index=['monthNum','monthName'],columns='job_skills',aggfunc='size',fill_value=0)
df_trendPivot = df_trendPivot.droplevel('monthNum')
df_trendPivot.loc['Total'] = df_trendPivot.sum()
df_trendPivot = df_trendPivot[df_trendPivot.loc['Total'].sort_values(ascending=False).index]
df_trendPivot = df_trendPivot.drop('Total')
df_trendPivot
```

```python
df_trendPivot.iloc[:,:5].plot(kind='line')

plt.title('Trending Top Skills for Data Analysts')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

#### Calculating the total jobs based on the month

```python
df_DATotals = df.groupby(['monthNum','monthName']).size()
df_DATotals.droplevel('monthNum')
```

#### Getting the percentage of jobs per skill by Month

```python
df_trendPivot_percent = df_trendPivot.iloc[0:12].div(df_DATotals/100,axis=0)
df_trendPivot_percent = df_trendPivot_percent.droplevel('monthNum')
df_trendPivot_percent
```

```python
from matplotlib.ticker import PercentFormatter

df_plot = df_trendPivot_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine() # remove top and right spines

plt.title('Trending Top Skills for Data Analysts')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

# annotate the plot with the top 5 skills using plt.text()
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i], color='black')

plt.show()
```
